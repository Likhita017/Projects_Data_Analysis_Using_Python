DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
F_col=['Juiciness','Crunchiness','Ripeness']
F=df[F_col]
T=df.Quality
```

```python
m=DecisionTreeClassifier()
m.fit(F,T)
```

```python
a=float(input("Enter juiciness of the apple"))
b=float(input("Enter Crunchiness of the apple"))
c=float(input("Enter Ripeness of the apple"))
pred=m.predict([[a,b,c]])
if pred=='good':
  print("Quality of apple is good")
else:
  print("Quality of apple is bad")
```

```python
from sklearn.tree import export_graphviz,plot_tree
plot_tree(m)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
F_col=['Juiciness','Crunchiness','Ripeness','Size']
F=df[F_col]
T=df.Quality
```

```python
M=RandomForestClassifier(n_estimators=10)
M.fit(F,T)
```

```python
a=float(input("Enter juiciness of the apple"))
b=float(input("Enter Crunchiness of the apple"))
c=float(input("Enter Ripeness of the apple"))
d=float(input("Enter size of the apple"))
pred=M.predict([[a,b,c,d]])
if pred == 'good':
  print("Quality of apple is good")
else:
  print("Quality of apple is bad")
```

```python
import os
output_dir='tree_visualizations'
os.makedirs(output_dir,exist_ok=True)
```

```python
M.estimators_
```

```python
for i, tree in enumerate(M.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=['Juiciness','Crunchiness','Ripeness','Size'],
                    class_names=[str(cls) for cls in M.classes_], filled=True, rounded=True)
```

```python
command= f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree{i} visualization saved to {tree_png_file}")
```
